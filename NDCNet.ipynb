```python
# Concatenate 1/2/5/7/8C/US06/SC04 discharging data for training
# and keep other datasets for testing
Real = np.concatenate((Real_1C,Real_2C),axis=0)
# Real = np.concatenate((Real,Real_3C),axis=0)
# Real = np.concatenate((Real,Real_4C),axis=0)
Real = np.concatenate((Real,Real_5C),axis=0)
# Real = np.concatenate((Real,Real_6C),axis=0)
Real = np.concatenate((Real,Real_7C),axis=0)
Real = np.concatenate((Real,Real_8C),axis=0)
# Real = np.concatenate((Real,Real_UDDS1),axis=0)
Real = np.concatenate((Real,Real_US061),axis=0)
# Real = np.concatenate((Real,Real_LA921),axis=0)
Real = np.concatenate((Real,Real_SC041),axis=0)

NDC = np.concatenate((NDC_1C,NDC_2C),axis=0)
# NDC = np.concatenate((NDC,NDC_3C),axis=0)
# NDC = np.concatenate((NDC,NDC_4C),axis=0)
NDC = np.concatenate((NDC,NDC_5C),axis=0)
# NDC = np.concatenate((NDC,NDC_6C),axis=0)
NDC = np.concatenate((NDC,NDC_7C),axis=0)
NDC = np.concatenate((NDC,NDC_8C),axis=0)
# NDC = np.concatenate((NDC,NDC_UDDS1),axis=0)
NDC = np.concatenate((NDC,NDC_US061),axis=0)
# NDC = np.concatenate((NDC,NDC_LA921),axis=0)
NDC = np.concatenate((NDC,NDC_SC041),axis=0)

Real = np.concatenate((Real,Real_1C_off),axis=0)
Real = np.concatenate((Real,Real_2C_off),axis=0)
# Real = np.concatenate((Real,Real_3C_off),axis=0)
# Real = np.concatenate((Real,Real_4C_off),axis=0)
Real = np.concatenate((Real,Real_5C_off),axis=0)
# Real = np.concatenate((Real,Real_6C_off),axis=0)
Real = np.concatenate((Real,Real_7C_off),axis=0)
Real = np.concatenate((Real,Real_8C_off),axis=0)
# Real = np.concatenate((Real,Real_UDDS2_off),axis=0)
Real = np.concatenate((Real,Real_US062_off),axis=0)
# Real = np.concatenate((Real,Real_LA922_off),axis=0)
Real = np.concatenate((Real,Real_SC042_off),axis=0)

NDC = np.concatenate((NDC,NDC_1C_off),axis=0)
NDC = np.concatenate((NDC,NDC_2C_off),axis=0)
# NDC = np.concatenate((NDC,NDC_3C_off),axis=0)
# NDC = np.concatenate((NDC,NDC_4C_off),axis=0)
NDC = np.concatenate((NDC,NDC_5C_off),axis=0)
# NDC = np.concatenate((NDC,NDC_6C_off),axis=0)
NDC = np.concatenate((NDC,NDC_7C_off),axis=0)
NDC = np.concatenate((NDC,NDC_8C_off),axis=0)
# NDC = np.concatenate((NDC,NDC_UDDS2_off),axis=0)
NDC = np.concatenate((NDC,NDC_US062_off),axis=0)
# NDC = np.concatenate((NDC,NDC_LA922_off),axis=0)
NDC = np.concatenate((NDC,NDC_SC042_off),axis=0)

print("Concatenate NDC train length: " + str(len(NDC)))
print("Concatenate Real train length: " + str(len(Real)))
```

```python
#Construct training data
trainX = np.zeros((len(NDC),5))
# trainX[:,0] = NDC[:,0] # 0: Vt
trainX[:,0] = NDC[:,1] # 1: Vb
trainX[:,1] = NDC[:,2] # 2: Vs
trainX[:,2] = NDC[:,3] # 3: V1
# trainX[:,0] = NDC[:,4] # 4: SoC
# trainX[:,3] = NDC[:,5] # 5: R0
trainX[:,3] = -NDC[:,6] # 6: Current
trainX[:,4] = NDC[:,7] # 7: temperature

# # voltage residual for NDCNet-1
# trainY = Real.reshape(len(Real),1) - NDC[:,0].reshape(len(NDC),1)
# # voltage for NDCNet-2
trainY = Real 
```

```python
scaler1 = MinMaxScaler()
scaler1.fit(trainX)
normalizedX = scaler1.transform(trainX)
scaler2 = MinMaxScaler()
scaler2.fit(trainY)
normalizedY = scaler2.transform(trainY)
trainX = normalizedX
trainY = normalizedY
```

```python
initScore = math.sqrt(mean_squared_error(Real.reshape(len(Real),1), NDC[:,0].reshape(len(NDC),1)))
print("Initial Score: RMSE " + str(initScore))
```

```python
model = Sequential()

model.add(Dense(32, input_dim=5, activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(1, activation='linear'))

model.compile(loss='mean_squared_error',optimizer='adam')
model.summary()
```

```python
# train NN
model.fit(trainX, trainY, epochs=200, verbose=1)
model.evaluate(trainX, trainY)
```

```python
# Training accuracy
trainPredict = model.predict(trainX)
trainPredict = scaler2.inverse_transform(trainPredict.reshape(len(trainPredict),1))
# trainScore = math.sqrt(mean_squared_error(Real, NDC[:,0].reshape(len(NDC),1) + trainPredict.reshape(len(trainPredict),1)))
trainScore = math.sqrt(mean_squared_error(Real, trainPredict.reshape(len(trainPredict),1)))
print("Train Score: RMSE " + str(trainScore))
```

```python
# Testing results
# Construct testing data for each current profile in the testing dataset
# The testing dataset includes 3/4/6C and UDDS/LA92 profiles
NDCtest = NDC_6C # 6C discharging with cooling fan on
NDCtest = np.concatenate((NDCtest,NDC_6C_off),axis=0) # 6C discharging with cooling fan off
Realtest = Real_6C # 6C discharging with cooling fan on
Realtest = np.concatenate((Realtest,Real_6C_off),axis=0) # 6C discharging with cooling fan off

testX = np.zeros((len(NDCtest),5))
# testX[:,0] = NDCTT[:,0] # 0: Vt
testX[:,0] = NDCtest[:,1] # 1: Vb
testX[:,1] = NDCtest[:,2] # 2: Vs
testX[:,2] = NDCtest[:,3] # 3: V1
# testX[:,0] = NDCtest[:,4] # 4: SoC
# testX[:,3] = NDCtest[:,5] # 5: R0
testX[:,3] = -NDCtest[:,6] # 6: Current
testX[:,4] = NDCtest[:,7] # 7: Temperature

# # voltage residual for NDCNet-1
# testY = Realtest.reshape(len(Realtest),1) - NDCtest[:,0].reshape(len(NDCtest),1)
# # predict voltage for NDCNet-2
testY = Realtest
```

```python
# Testing accuracy
normalizedXtest = scaler1.transform(testX)
testPredict = model.predict(normalizedXtest)
testPredict = scaler2.inverse_transform(testPredict.reshape(len(testPredict),1))
testScore = math.sqrt(mean_squared_error(Realtest.reshape(len(testPredict),1), testPredict.reshape(len(testPredict),1)))
print("Test Score: RMSE " + str(testScore))
InitialScore = math.sqrt(mean_squared_error(Realtest, NDCtest[:,0]))
print("Initial Score: RMSE " + str(InitialScore))
```
